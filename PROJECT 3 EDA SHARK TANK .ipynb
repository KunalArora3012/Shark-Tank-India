```python
import seaborn as sns
import matplotlib.pyplot as plt

# create data
categories = ["Min0", "Max", "Mean"]
Pitcher_Ask_Amount = [5,30000, 320,]
Ask_Equity = [0.25, 25, 5,]
Ask_Valuation = [50, 120000,3852]

# set figure size
plt.figure(figsize=(10, 5))

# create 3 subplots
plt.subplot(1, 3, 1)
sns.barplot(x=categories, y=Pitcher_Ask_Amount)
plt.title("Plot 1")

plt.subplot(1, 3, 2)
sns.barplot(x=categories, y=Ask_Equity)
plt.title("Plot 2")

plt.subplot(1, 3, 3)
sns.barplot(x=categories, y=Ask_Valuation)
plt.title("Plot 3")

# adjust spacing between subplots
plt.tight_layout()

# show plot
plt.show()
```
